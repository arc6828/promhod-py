```python
from bs4 import BeautifulSoup
import requests
import json,codecs
import pandas as pd

# Step 1: Open and read the HTML file
html_file_path = "data/flood-index-thairath.html"  # Replace with your HTML file path
with open(html_file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# Step 2: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find the <main> tag
# main_tag = soup.find("main")

# Find the <div> tag with class "container"
container_divs = soup.find_all("div", class_="container")

# Find all <a> tags
a_tags = []
for container_div in container_divs:
    a_tags += container_div.find_all("a")

# Extract and print URLs
data = []
for idx, tag in enumerate(a_tags, 1):
    # title
    title = tag.text  # Use text to get string without HTML tags
    title = title.replace("\n", " ").replace("  ", " ")  # clean
    title = title.replace("    ", " ")  # clean
    title = title.replace("  ", " ")  # clean
    title = title.strip()  # clean
    # title2 = tag.decode_contents()

    # url
    url = tag.get("href")  # Use .get() to avoid errors if href is missing
    absolute_url = url
    domain = "https://www.thairath.co.th"
    if url:
        # Check if the URL has a protocol; if not, add "http://"
        if not url.startswith(("http://", "https://")):
            absolute_url = f"{domain}{url}"
        print(f"Link {idx}: {absolute_url}")
        # processed_links.append(url)
    if domain not in absolute_url or title == "":
        continue

    # tags กรององค์ประกอบสุดท้ายออกถ้าหากเป็นตัวเลข        
    tags = absolute_url.replace(f"{domain}/", "").split("/")
    tags = tags[:-1] if isinstance(tags[-1], (int, float)) else tags
    blacklist_tags = ["tags","sport","entertain","horoscope","lifestyle","money"]
    flag = True
    for t in tags:
        if t in blacklist_tags:
            flag = False
            break
    if flag == False: 
        continue

    # publish_date
    # sibling of parent
    span_date = tag.parent.parent.find('span')
    publish_date = ""
    if span_date is not None:
        publish_date = span_date.text

    item = {
        "title" : title,
        # "title2":title2,
        "url" : absolute_url,
        "tags" : tags,
        "publish_date" : publish_date,
    }
    data.append(item)

filename = "data/flood-index-thairath.json"
with codecs.open(filename, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False)
    print(f"Write file {filename}")

df = pd.DataFrame(data)
filename = "data/flood-index-thairath.xlsx"
df.to_excel(filename, index=False)  # บันทึกเป็น Excel
print("print big excel")

    
```
